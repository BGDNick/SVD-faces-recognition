```python
from recognition import SVDrecognition
import matplotlib.pyplot as plt
import numpy as np
```

```python
def quality(num_vecs_):
    rec = SVDrecognition()

    rec.fit(num_vecs=num_vecs_)
    corr = 0
    for i in range(14):
        img = plt.imread(f'./test/{i + 1}_3.jpg')

        _, name = rec.predict(img)

        if name == f'./train/{i + 1}_1.jpg' or name == f'./train/{i + 1}_2.jpg':
            # print("correct", i)
            corr += 1

    return corr / 14
```

```python
quality(10)
```

```python
quality(20)
```

## Now, let's look at some good examples

```python
def show_prediction(num_vecs_, img):
    rec = SVDrecognition()
    rec.fit(num_vecs=num_vecs_)
    _, name = rec.predict(img)

    name = plt.imread(name)

    plt.subplot(121)
    plt.imshow(img)
    plt.title("test image")
    plt.subplot(122)
    plt.imshow(name)
    plt.title("reference image")
```

```python
img = plt.imread(f'./test/11_3.jpg')

show_prediction(20, img)
```

```python
img = plt.imread(f'./test/2_3.jpg')

show_prediction(20, img)
```
